```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from app.data_cleaning import clean_loan_data

# Load dataset
df_raw = pd.read_csv(r'C:\Users\91825\credit_risk_chat\data\loan_data.csv')

# Clean dataset using your cleaning function
df = clean_loan_data(df_raw)

# Quick info
print("âœ… Cleaned dataset loaded successfully!")
df.info()
```

```python
print("Dataset shape:", df.shape)
print("\nMissing values per column:\n", df.isna().sum())

display(df.head())
display(df.describe(include='all').transpose())
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='loan_status', data=df, palette='Set2')
plt.title("Loan Status Distribution (0 = Default, 1 = Paid)")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()

default_rate = (df['loan_status'] == 0).mean()
print(f"ðŸ“‰ Default Rate: {default_rate*100:.2f}%")
```

```python
num_cols = ['person_income', 'loan_amnt', 'loan_int_rate', 'credit_score']
fig, axes = plt.subplots(2, 2, figsize=(12,8))

for col, ax in zip(num_cols, axes.flatten()):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f"Distribution of {col}")

plt.tight_layout()
plt.show()
```

```python
cat_cols = ['person_gender', 'person_education', 'person_home_ownership', 'loan_intent']

fig, axes = plt.subplots(2, 2, figsize=(12,10))

for col, ax in zip(cat_cols, axes.flatten()):
    sns.countplot(y=col, data=df, ax=ax, order=df[col].value_counts().index)
    ax.set_title(f"Distribution of {col}")
    ax.set_xlabel("Count")

plt.tight_layout()
plt.show()
```

```python
for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.barplot(x=col, y='loan_status', data=df, estimator=lambda x: 1 - np.mean(x))
    plt.title(f"Default Rate by {col}")
    plt.ylabel("Default Rate")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='credit_score', y='person_income', hue='loan_status', data=df, alpha=0.6)
plt.title("Credit Score vs Income (Colored by Loan Status)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

```

```python
segment_df = df.groupby('loan_intent').agg({
    'loan_amnt': 'mean',
    'loan_int_rate': 'mean',
    'loan_status': lambda x: 1 - x.mean()
}).reset_index()

segment_df.rename(columns={'loan_status': 'default_rate'}, inplace=True)

plt.figure(figsize=(8,5))
sns.barplot(x='loan_intent', y='default_rate', data=segment_df, palette='viridis')
plt.title("Average Default Rate by Loan Purpose")
plt.ylabel("Default Rate")
plt.xticks(rotation=45)
plt.show()

segment_df
```

```python
print("ðŸ“ˆ Key Insights:")
print("- Average default rate:", round(default_rate*100, 2), "%")
print("- Highest default rate segment:", segment_df.loc[segment_df['default_rate'].idxmax(), 'loan_intent'])
print("- Strongest correlation with default: credit_score (negative)")
```
